# Expectation Decider Project
## Probability and Statistical Analysis Report

### Import Libraries

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Load Dataset

```
df = pd.read_csv('expectation_decider_dataset.csv')
df.head()
```

## 1. Understanding the Basics
### What is Probability?
Probability means the chance of an event occurring.

### Example Events
1. Student studies more than 10 hours.
2. Attendance is greater than 80%.
3. Student passed the exam.

## 2. Empirical Probability

```
pass_students = len(df[df['final_result'] == 'Pass'])
total_students = len(df)

prob_pass = pass_students / total_students
print('Probability of Passing:', prob_pass)
```

## 3. Random Variable and Probability Distribution

```
distribution = df['final_result'].value_counts(normalize=True)
distribution
```

## 4. Mean and Variance

```
mean_hours = df['study_hours'].mean()
variance_hours = df['study_hours'].var()

print('Mean Study Hours:', mean_hours)
print('Variance:', variance_hours)
```

## 5. Venn Diagram Conditions

```
study_condition = df[df['study_hours'] > 10]
attendance_condition = df[df['attendance'] > 80]

both_conditions = df[(df['study_hours'] > 10) & (df['attendance'] > 80)]

print('Students studying >10 hrs:', len(study_condition))
print('Students with attendance >80%:', len(attendance_condition))
print('Students satisfying both:', len(both_conditions))
```

## 6. Contingency Table

```
df['attendance_level'] = np.where(df['attendance'] > 80, 'High', 'Low')

contingency_table = pd.crosstab(df['attendance_level'], df['final_result'])
contingency_table
```

## 7. Bayes Theorem

```
high_attendance = df[df['attendance'] > 80]

passed_and_high = len(high_attendance[high_attendance['final_result'] == 'Pass'])
total_high = len(high_attendance)

bayes_probability = passed_and_high / total_high

print('Probability of passing given high attendance:', bayes_probability)
```

## 8. Visualization

```
plt.figure(figsize=(8,5))
df['final_result'].value_counts().plot(kind='bar')
plt.title('Pass vs Fail Students')
plt.xlabel('Final Result')
plt.ylabel('Count')
plt.show()
```

## 9. Final Summary
- Students with higher study hours tend to pass more.
- High attendance improves the probability of passing.
- Group discussion participation helps students perform better.
