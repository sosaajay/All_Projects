# 📊 Step 1: Measures of Central Tendency and Dispersion

## 🎯 Objective

In this step, the basic statistical measures are calculated using the student dataset. These measures help us understand the average performance and the spread of students' scores.

The following statistical measures are calculated:

- Mean of **Math_Score**
- Median of **Math_Score**
- Mode of **Math_Score**
- Range of **Science_Score**
- Variance of **Science_Score**
- Standard Deviation of **Science_Score**

```python
mean = df["Math_Score"].mean()
median = df["Math_Score"].median()
mode = df["Math_Score"].mode()[0]

range_value = df["Science_Score"].max() - df["Science_Score"].min()
variance = df["Science_Score"].var()
std = df["Science_Score"].std()

print("Mean of Math Score :", mean)
print("Median of Math Score :", median)
print("Mode of Math Score :", mode)
print("Range of Science Score :", range_value)
print("Variance of Science Score :", variance)
print("Standard Deviation of Science Score :", std)
```

## ✅ Conclusion

The Mean, Median, and Mode describe the central tendency of the Math Score data, while the Range, Variance, and Standard Deviation describe the variability of the Science Score data. These statistical measures help summarize the dataset and provide a better understanding of students' academic performance.

# 🎲 Step 2: Probability Basics

## 🎯 Objective

The objective of this step is to calculate the probability of students passing the examination and analyze the relationship between study hours and passing status.

The following tasks are performed:

- Calculate the probability of students passing.
- Create a contingency table between **Pass_Fail** and **Hours_Studied > 5**.
- Calculate the conditional probability of passing when the study hours are greater than 5.

```python
probability = len(df[df["Pass_Fail"] == 1]) / len(df)

print("Probability of Passing :", probability)

table = pd.crosstab(df["Pass_Fail"], df["Hours_Studied"] > 5)

print("\nContingency Table\n")
print(table)

conditional_probability = len(df[(df["Pass_Fail"] == 1) & (df["Hours_Studied"] > 5)]) / len(df[df["Hours_Studied"] > 5])

print("\nConditional Probability P(Pass | Hours_Studied > 5) :", conditional_probability)
```

## ✅ Conclusion

The probability of passing shows the proportion of students who successfully passed the examination. The contingency table summarizes the relationship between study hours and pass/fail status. The conditional probability indicates the likelihood of passing for students who studied more than 5 hours, helping us understand the impact of study time on academic performance.

# 📊 Step 3: Distribution and Visualization

## 🎯 Objective

The objective of this step is to analyze the distribution of student scores using statistical measures and graphical visualization.

The following tasks are performed:

- Plot a Histogram with Normal Distribution Curve for **Math_Score**.
- Calculate **Skewness** and **Kurtosis** for **Science_Score**.
- Create a **Q-Q Plot** for **English_Score** to check whether the data follows a normal distribution.

```python
print("Skewness :", stats.skew(df["Science_Score"]))

print("Kurtosis :", stats.kurtosis(df["Science_Score"]))
```

```python
plt.figure(figsize=(8,5))

sns.histplot(df["Math_Score"], kde=True)

plt.title("Histogram of Math Score")

plt.xlabel("Math Score")

plt.ylabel("Frequency")

plt.show()
```

```python
stats.probplot(df["English_Score"], dist="norm", plot=plt)

plt.title("Q-Q Plot of English Score")

plt.show()
```

## ✅ Conclusion

The histogram helps us understand the distribution of **Math_Score**, while the **Q-Q Plot** is used to compare the distribution of **English_Score** with a normal distribution. The calculated **Skewness** and **Kurtosis** describe the shape and spread of the **Science_Score** data, making it easier to identify symmetry and the presence of outliers.

# ➗ Step 4: Linear Algebra Mini Task

## 🎯 Objective

The objective of this step is to perform basic Linear Algebra operations using the first five students' Math Score and Science Score.

The following operations are performed:

- Represent Math Score and Science Score as vectors.
- Calculate the Dot Product.
- Calculate the L1 Norm.
- Calculate the L2 Norm.
- Calculate the Angle Between the Two Vectors.

```python
math_vector = df["Math_Score"][:5].values
science_vector = df["Science_Score"][:5].values

print("Math Vector")
print(math_vector)

print("\nScience Vector")
print(science_vector)

dot_product = np.dot(math_vector, science_vector)

l1_norm = np.linalg.norm(math_vector, 1)

l2_norm = np.linalg.norm(math_vector, 2)

angle = np.degrees(np.arccos(dot_product / (np.linalg.norm(math_vector) * np.linalg.norm(science_vector))))

print("\nDot Product :", dot_product)

print("L1 Norm :", l1_norm)

print("L2 Norm :", l2_norm)

print("Angle Between Two Vectors :", angle)
```

## ✅ Conclusion

The Linear Algebra operations were successfully performed using the first five students' Math Score and Science Score. The Dot Product measures the relationship between the two vectors, the L1 and L2 Norms measure the vector lengths, and the Angle Between the Vectors indicates the similarity in their directions.
