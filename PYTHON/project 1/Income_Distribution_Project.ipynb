
# Analyzing Income Distribution and Household Demographics

## Objective
This project analyzes household demographic and income data using Python.

The notebook includes:
- Types of Data
- Central Tendency
- Measures of Dispersion
- Distribution Analysis
- Visualization
- Skewness and Kurtosis
- Data Categorization

Dataset Size: **200 Rows**


```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew, kurtosis, norm

df = pd.read_csv('household_income_dataset_200_rows.csv')

df.head()
```


# 1. Types of Data

### Categorical Columns
- Household_ID
- Education_Level
- Owns_House
- Urban_Rural

### Numerical Columns
- Age_of_Household_Head
- Household_Income
- Family_Size



# 2. Central Tendency

Calculate Mean, Median and Mode of:
- Household Income
- Age of Household Head


```python

income_mean = df['Household_Income'].mean()
income_median = df['Household_Income'].median()
income_mode = df['Household_Income'].mode()[0]


age_mean = df['Age_of_Household_Head'].mean()
age_median = df['Age_of_Household_Head'].median()
age_mode = df['Age_of_Household_Head'].mode()[0]

print("Income Mean:", income_mean)
print("Income Median:", income_median)
print("Income Mode:", income_mode)

print("\nAge Mean:", age_mean)
print("Age Median:", age_median)
print("Age Mode:", age_mode)
```


# 3. Measures of Dispersion

This section calculates:
- Range
- Variance
- Standard Deviation
- IQR


```python

income_range = df['Household_Income'].max() - df['Household_Income'].min()
income_variance = df['Household_Income'].var()
income_std = df['Household_Income'].std()

Q1 = df['Household_Income'].quantile(0.25)
Q3 = df['Household_Income'].quantile(0.75)
IQR = Q3 - Q1

print("Income Range:", income_range)
print("Variance:", income_variance)
print("Standard Deviation:", income_std)
print("IQR:", IQR)
```


# 4. Distribution Analysis

This section includes:
- Histogram
- Gaussian Distribution Curve
- Skewness
- Kurtosis


```python

plt.figure(figsize=(8,5))
sns.histplot(df['Household_Income'], kde=True)

mu, std = norm.fit(df['Household_Income'])

xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)

plt.plot(x, p, linewidth=2)
plt.title('Histogram and Gaussian Distribution')
plt.xlabel('Household Income')
plt.ylabel('Frequency')
plt.show()
```

```python

income_skewness = skew(df['Household_Income'])
income_kurtosis = kurtosis(df['Household_Income'])

print("Skewness:", income_skewness)
print("Kurtosis:", income_kurtosis)
```


# 5. Visualizations


```python

plt.figure(figsize=(8,5))
sns.boxplot(x='Education_Level', y='Household_Income', data=df)
plt.title('Household Income by Education Level')
plt.xticks(rotation=10)
plt.show()
```

```python

plt.figure(figsize=(8,5))
sns.scatterplot(x='Age_of_Household_Head', y='Household_Income', hue='Urban_Rural', data=df)
plt.title('Age vs Income')
plt.show()
```


# 6. Conclusion

### Findings
- Most households fall within a moderate income range.
- Income distribution is approximately normal.
- Higher education levels generally show higher income.
- Urban households tend to have slightly higher income patterns.
