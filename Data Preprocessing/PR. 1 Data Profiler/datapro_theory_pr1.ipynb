# 🟦 Part A: Fundamentals

## 1️⃣ What is Data Analysis? 📈

Data Analysis is the process of collecting, organizing, cleaning, transforming, and analyzing raw data to extract useful information and meaningful patterns. It helps in understanding data behavior and supports better decision-making.

In data analysis, different techniques like statistical analysis, visualization, and data exploration are used to identify trends, relationships, and important insights from the dataset.

---

# 2️⃣ Planning a Data Science Project 🚀

A Data Science project is planned by following a proper workflow that converts a business problem into a data-driven solution. It includes collecting data, processing it, analyzing it, building models, and evaluating results.

The main steps of a Data Science project are:

### 🎯 Problem Definition
In this step, the objective of the project is identified. The problem statement defines what we want to predict or analyze.

### 📥 Data Collection
Data is collected from different sources such as CSV files, JSON files, SQL databases, and APIs. Good quality data is important for accurate results.

### 🧹 Data Cleaning
Raw data may contain missing values, duplicate records, or incorrect formats. Data cleaning removes these problems and improves data quality.

### 📊 Exploratory Data Analysis (EDA)
EDA is used to understand the structure and behavior of data using statistical methods and visualizations like graphs and charts.

### ⚙️ Feature Engineering
Feature engineering converts raw data into useful features that can be understood by machine learning algorithms.

### 🤖 Model Building
Machine learning algorithms are applied to create a model that can learn patterns from historical data and make predictions.

### ✅ Model Evaluation
The performance of the model is tested using different evaluation techniques to check accuracy and reliability.

# 3️⃣ Machine Learning Problem Statement 🤖

The objective of this project is to predict whether a customer will churn or not based on their purchase behavior. Customer churn means whether a customer stops using a product or service.

The model uses customer information like age, gender, purchase amount, previous purchases, subscription status, and review rating as input features.

The output variable is Churn, where 0 represents that the customer stays and 1 represents that the customer leaves. This is a classification problem because the output is divided into categories.

# 4️⃣ Tensor and its Types 🔢

A Tensor is a multi-dimensional data structure used in machine learning and deep learning to represent numerical data. Tensors help computers process large amounts of structured data efficiently.

### 1D Tensor
A one-dimensional tensor contains a single list of values. It is used to represent simple data like a sequence of numbers.

Example:
[1,2,3,4]

### 2D Tensor
A two-dimensional tensor represents data in the form of rows and columns, similar to a matrix.

Example:
[[1,2],[3,4]]

### 3D Tensor
A three-dimensional tensor contains multiple matrices and is commonly used in image processing and deep learning applications.

# 🟩 Part B: Data Acquisition 📥

Data Acquisition is the process of collecting data from different sources and preparing it for analysis. Data can be collected from files, databases, or online services.

In this project, data is acquired from four different sources: CSV, JSON, SQL Database, and API.

---

## 📄 CSV Data Source

CSV (Comma Separated Values) is a file format used to store data in rows and columns. It is commonly used because it is simple, lightweight, and can easily be loaded using Python libraries like Pandas.

---

## 🗂️ JSON Data Source

JSON (JavaScript Object Notation) is a format used to store and exchange data using key-value pairs. It is widely used in web applications and APIs because it is easy for machines and humans to read.

---

## 🛢️ SQL Database

SQL databases store structured data in tables consisting of rows and columns. SQL allows users to store, manage, search, and retrieve large amounts of data using queries.

---

## 🌐 API Data Source

API (Application Programming Interface) allows different software applications to communicate and exchange data. APIs are commonly used to fetch real-time data from online services.

# 🟨 Part C: Data Exploration and Understanding 🔍

Data Exploration is the process of studying and understanding the dataset before performing analysis. It helps identify data structure, patterns, and possible issues.

Important functions used:

### head()
Shows the first few rows of the dataset to understand the type of information available.

### info()
Provides information about columns, data types, and missing values present in the dataset.

### describe()
Generates statistical summaries such as mean, minimum, maximum, and standard deviation of numerical columns.

---

# 🟥 Part D: Data Cleaning 🧹

Data Cleaning is an important step where incorrect, incomplete, or duplicate data is corrected before analysis.

Missing values are handled by removing them or replacing them with suitable values. Duplicate records are removed to avoid incorrect results.

Data types are also corrected, such as converting text dates into proper date format.

---

# 🟪 Part E: Exploratory Data Analysis (EDA) 📊

Exploratory Data Analysis is used to discover patterns, relationships, and trends in data through statistics and visualization.

### Univariate Analysis
It focuses on analyzing a single variable. Example: Studying customer age distribution.

### Bivariate Analysis
It studies the relationship between two variables. Example: Purchase amount and churn relationship.

### Multivariate Analysis
It analyzes multiple variables together to find complex relationships using techniques like correlation analysis.

---

# 🟧 Part F: Data Profiling 📑

Data Profiling is the process of creating a detailed summary of a dataset. It helps understand the quality, structure, and characteristics of data.

A profiling report includes information about missing values, duplicate records, data types, statistical summary, and relationships between variables.

In this project, the ydata-profiling library is used to automatically generate a complete data analysis report.

Data profiling and analysis help convert raw customer data into meaningful insights. By cleaning, exploring, and analyzing data, we can understand customer behavior and prepare the dataset for machine learning models. 🚀
