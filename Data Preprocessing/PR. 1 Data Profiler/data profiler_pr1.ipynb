# 📊 PART D - EXPLORATORY DATA ANALYSIS (EDA)

```python
# Univariate Analysis - Age

plt.figure(figsize=(6,4))
sns.histplot(df["Age"], kde=True)

plt.title("Age Distribution")
plt.show()
```

```python
# Univariate Analysis - Purchase Amount

plt.figure(figsize=(6,4))
sns.histplot(df["Purchase Amount (USD)"], kde=True)

plt.title("Purchase Amount Distribution")
plt.show()
```

```python
# Churn Distribution

sns.countplot(x="Churn", data=df)

plt.title("Churn Distribution")
plt.show()
```

```python
# Gender vs Churn

sns.countplot(
    x="Gender",
    hue="Churn",
    data=df
)

plt.title("Gender vs Churn")
plt.show()
```

```python
# Category vs Churn

sns.countplot(
    x="Category",
    hue="Churn",
    data=df
)

plt.xticks(rotation=45)
plt.show()
```

```python
# Purchase Amount vs Churn

sns.boxplot(
    x="Churn",
    y="Purchase Amount (USD)",
    data=df
)

plt.show()
```

```python
# Multivariate Analysis

numeric_data = df.select_dtypes(include="number")

plt.figure(figsize=(10,6))

sns.heatmap(
    numeric_data.corr(),
    annot=True
)

plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.pairplot(
    df[
        [
            "Age",
            "Purchase Amount (USD)",
            "Previous Purchases",
            "Review Rating",
            "Churn"
        ]
    ]
)

plt.show()
```

# 📑 PART E - DATA PROFILING

```python
# Convert columns into numeric format

df["Purchase Amount (USD)"] = pd.to_numeric(
    df["Purchase Amount (USD)"],
    errors="coerce"
)

df["Previous Purchases"] = pd.to_numeric(
    df["Previous Purchases"],
    errors="coerce"
)


df.dtypes
```

```python
# Feature Engineering

df["Purchase_Per_History"] = (
    df["Purchase Amount (USD)"] /
    (df["Previous Purchases"] + 1)
)

df.head()
```
