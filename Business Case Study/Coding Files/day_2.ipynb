```python
import numpy as np
import pandas as pd
```

### Measure of Central Tendancy

```python
# Mean
# Median
# Mode
```

```python
data = [10, 20, 30, 40, 50]
```

```python
data
```

```python
# Mean

mean = sum(data) / len(data)
mean
```

```python
arr = np.array(data)
arr
```

```python
print(arr.mean())
```

```python
s = pd.Series(data)
s
```

```python
print(s.mean())
```

```python
data
```

```python
# Median

median = data[len(data) // 2]
median
```

```python
s.median()
```

```python
# Mode

data2 = [10, 20, 20, 20, 30, 30, 30, 40]
data2
```

```python
arr2 = pd.Series(data2)
arr2
```

```python
arr2.mode()
```

```python
mode = arr2.mode()[0]
print(mode)
```

```python
import statistics
```

```python
mode = statistics.mode(data2)
mode
```

### Measure of Dispersion

```python
# Range
# Variance
# Standard Deviation
```

```python
data
```

```python
# Range

range = max(data) - min(data)
range
```

```python
# Variance

data
```

```python
arr = np.array(data)
arr
```

```python
var = arr.var()
print(var)
```

```python
# Standard Deviation

std = var ** 0.5
print(std)
```

```python
std = arr.std()
print(std)
```
