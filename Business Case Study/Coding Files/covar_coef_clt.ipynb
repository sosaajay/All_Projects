```python
import numpy as np

# Define two sample datasets
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])

# Calculate covariance manually
mean_x = np.mean(x)
mean_y = np.mean(y)
cov_xy = np.sum((x - mean_x) * (y - mean_y)) / (len(x) - 1)
print(f"Covariance between x and y: {cov_xy}")
```

```python
# Calculate Pearson correlation coefficient
std_x = np.std(x, ddof=1)
std_y = np.std(y, ddof=1)
pearson_r = cov_xy / (std_x * std_y)
print(f"Pearson correlation: {pearson_r}")
```

```python
import scipy.stats as stats

# Calculate Spearman rank correlation
spearman_r, p_value = stats.spearmanr(x, y)
print(f"Spearman correlation: {spearman_r}")
```

```python
import matplotlib.pyplot as plt

# Generate multiple samples and calculate their means
sample_size = 30
num_samples = 1000
sample_means = []

# Take random samples from a uniform distribution
for _ in range(num_samples):
    sample = np.random.uniform(0, 10, sample_size)
    sample_means.append(np.mean(sample))

# Plot the distribution of sample means
plt.figure(figsize=(10, 6))
plt.hist(sample_means, bins=30, density=True, alpha=0.7, edgecolor='black')
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.title('Central Limit Theorem: Distribution of Sample Means')
plt.axvline(np.mean(sample_means), color='red', linestyle='--', label='Mean of sample means')
plt.legend()
plt.show()

print(f"Mean of sample means: {np.mean(sample_means)}")
print(f"Standard deviation of sample means: {np.std(sample_means)}")
```

```python
import numpy as np
from scipy.stats import norm, expon, poisson

import matplotlib.pyplot as plt

# Define x values
x = np.linspace(0, 10, 1000)

# Normal distribution
plt.subplot(2, 2, 1)
plt.plot(x, norm.pdf(x, loc=5, scale=1), label='Normal')
plt.title('Normal Distribution')
plt.legend()

# Exponential distribution
plt.subplot(2, 2, 2)
plt.plot(x, expon.pdf(x, scale=1), label='Exponential')
plt.title('Exponential Distribution')
plt.legend()

# Poisson distribution (PMF)
plt.subplot(2, 2, 3)
x_poisson = np.arange(0, 10)
plt.bar(x_poisson, poisson.pmf(x_poisson, mu=3), label='Poisson')
plt.title('Poisson Distribution')
plt.legend()

# Uniform distribution
plt.subplot(2, 2, 4)
plt.plot(x, np.ones_like(x)/10, label='Uniform')
plt.title('Uniform Distribution')
plt.legend()

plt.tight_layout()
plt.show()
```
