```python
import numpy as np
from scipy import stats
```

### One-sample z-test

##### Q.1 The average heights of all reidents in a city is 168 cm with a population standard deviation of 3.9. A doctor believe the mean to be different. He measured the highet of 36 individuals and found the average to be 169.5 cm.

```python
population_mean = 168  # National average height in cm
population_std = 3.9   # Known standard deviation
sample_size = 36       # Sample size
sample_mean = 169.5    # Sample mean height in cm
alpha=0.05

# Calculate the critical z-value for a two-tailed test at alpha=0.05

critical_z = stats.norm.ppf(1-(alpha/2))
print(f"Critical Z-Value (Decision Boundary): ±{critical_z:.2f}")

# Perform one-sample z-test
# H0: sample mean = population mean (no difference)
# H1: sample mean != population mean (there is a difference)
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))  # Two-tailed test

# Print results
print(f"Sample Mean: {sample_mean:.2f} cm")
print(f"Z-Score: {z_score:.2f}")
print(f"P-Value: {p_value:.4f}")

# Method-1
# if p_value < 0.05:
#     print("The p-value is less than 0.05, so we reject the null hypothesis. The sample height is significantly different from the population average.")
# else:
#     print("The p-value is greater than or equal to 0.05, so we fail to reject the null hypothesis. The sample height is not significantly different from the population average.")

# Method-2
if (z_score > critical_z) or (z_score < -critical_z):
    print("The z_score is falls outside of Decision Boundary, so we reject the null hypothesis. The sample height is significantly different from the population average.")
else:
    print("The z_score is falls within the Decision Boundary, so we fail to reject the null hypothesis. The sample height is not significantly different from the population average.")
```

### One-sample t-test

##### Q.1 In the population the average IQ is 100. A team of researchers want to test a new medication to see if it has either a positive or negative effect on intelligence, or no effect at all. A sample of 30 participants who have taken the medication has a mean of 140 with a Standard deviation of 20. Did the medication affect intelligence? CI=95%

```python
# One-sample t-test for IQ medication effect
# Population mean IQ: 100
# Sample size: 30
# Sample mean IQ: 140
# Sample standard deviation: 20
# Confidence interval: 95% (alpha = 0.05)

population_mean = 100  # Known population mean IQ
sample_size = 30       # Sample size
sample_mean = 140      # Sample mean IQ
sample_std = 20        # Sample standard deviation
alpha = 0.05           # Significance level for 95% CI

# Calculate the critical t-value for a two-tailed test at alpha=0.05
critical_t = stats.t.ppf(1 - (alpha / 2), df=sample_size - 1)
print(f"Critical T-Value (Decision Boundary): ±{critical_t:.2f}")

# Perform one-sample t-test
# H0: sample mean = population mean (no effect)
# H1: sample mean != population mean (positive or negative effect)
t_score = (sample_mean - population_mean) / (sample_std / np.sqrt(sample_size))
p_value = 2 * (1 - stats.t.cdf(abs(t_score), df=sample_size - 1))  # Two-tailed test

# Print results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"T-Score: {t_score:.2f}")
print(f"P-Value: {p_value:.4f}")

# Method-1: Decision based on p-value
# if p_value < alpha:
#     print("The p-value is less than 0.05, so we reject the null hypothesis. The medication significantly affects intelligence (positive or negative effect).")
# else:
#     print("The p-value is greater than or equal to 0.05, so we fail to reject the null hypothesis. There is no significant effect of the medication on intelligence.")

# Method-2: Alternative decision based on t-score
if (t_score > critical_t) or (t_score < -critical_t):
    print("The t-score falls outside the decision boundary, so we reject the null hypothesis. The medication significantly affects intelligence.")
else:
    print("The t-score falls within the decision boundary, so we fail to reject the null hypothesis. There is no significant effect.")
```
